```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("survey.csv")
df=df.drop(['Timestamp'], axis=1)
print("Shape of df - {}".format(df.shape))
print("columns")
print(df.columns)


#function to display pie charts 
def create_pie_chart(df, column):
    df3 = pd.DataFrame(df[column].value_counts().reset_index())
    labels=df[column].value_counts().index.tolist()
    #Assign it to another variable 

    plt.style.use('ggplot')
    plt.figure().set_facecolor('white')
    #Set bg white
    plt.title("Responses for "+ "'"+str(column)+"'")


    plt.pie(data=df3,x=column,autopct='%1.0f%%',labels=labels)
    #Pie chart functions

    plt.show()
    
    
df.head()
```

```python
#Filter people who are only in Chennai and play games and drop the columns
create_pie_chart(df,'Do you play video games?')
df=df[df['Do you play video games?']=='Yes']
df.drop(['Do you play video games?'],inplace=True,axis=1)
create_pie_chart(df,'Are you from Chennai or ever resided in Chennai?')
df=df[df['Are you from Chennai or ever resided in Chennai?']=='Yes']
df.drop(['Are you from Chennai or ever resided in Chennai?'],inplace=True,axis=1)
df.head()
```

```python

#Cannot fit in a pie chart
df.drop(['5. What is genre you prefer for gaming?'],inplace=True,axis=1)
df.drop(['21. Are you open to gaming as a career? Why so?'],inplace=True,axis=1)

print("\nPie charts created for data who are in Chennai and Play games\n")
for cols in df.columns:
    create_pie_chart(df,cols)
```
